```
import numpy as np
```

### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
array = np.random.randint(1, 51, size=10)

media = np.mean(array)
desvio_padrao = np.std(array)

print("Array: ", array)
print("Média: ", media)
print("Desvio padrão: ", desvio_padrao)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
array = np.random.rand(10)
raiz_quadrada = np.sqrt(array[array <= 0.5])

print("Array: ", array)
print("Raiz quadrada dos valores menores ou iguais a 0,5: ", raiz_quadrada)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)

for i in range(3):
    print("Máximo da linha ", i+1, " de cada matriz 2D:")
    print(np.max(matriz[i], axis=1))
    print("-------------------------")
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5, 5)
media_linhas = np.mean(matriz, axis=1)
matriz_subtraida = matriz - media_linhas

print("Matriz original:")
print(matriz)
print("-------------------------")
print("Média de cada linha:")
print(media_linhas)
print("-------------------------")
print("Matriz com a média de cada linha subtraída:")
print(matriz_subtraida)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4, 4)
diagonal_principal = np.diag(matriz)
diagonal_secundaria = np.diag(np.fliplr(matriz))

media_diagonal_principal = np.mean(diagonal_principal)
media_diagonal_secundaria = np.mean(diagonal_secundaria)

desvio_padrao_diagonal_principal = np.std(diagonal_principal)
desvio_padrao_diagonal_secundaria = np.std(diagonal_secundaria)

print("Matriz:")
print(matriz)
print("-------------------------")
print("Diagonal principal:")
print(diagonal_principal)
print("-------------------------")
print("Diagonal secundária:")
print(diagonal_secundaria)
print("-------------------------")
print("Média da diagonal principal:", media_diagonal_principal)
print("Média da diagonal secundária:", media_diagonal_secundaria)
print("-------------------------")
print("Desvio padrão da diagonal principal:", desvio_padrao_diagonal_principal)
print("Desvio padrão da diagonal secundária:", desvio_padrao_diagonal_secundaria)
```
