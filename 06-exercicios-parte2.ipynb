### 11. Crie um programa em Python que receba uma lista de números inteiros como entrada e retorne uma nova lista contendo apenas os números que aparecem um número ímpar de vezes.

```python
def numeros_impares(lista):
    """
    Recebe uma lista de números inteiros e retorna uma nova lista
    contendo apenas os números que aparecem um número ímpar de vezes.
    """
    numeros = set(lista) # obtém um conjunto com os números únicos da lista
    impares = [x for x in numeros if lista.count(x) % 2 == 1] # verifica a contagem de cada número e adiciona os ímpares à lista resultante
    return impares
```

```python
lista = [1, 2, 3, 4, 5, 6, 7, 8, 9, 9, 9, 8, 7, 6, 5, 4, 3, 2, 1]
impares = numeros_impares(lista)
print(impares) 
```

### 12. Crie um programa em Python que receba duas listas de números inteiros como entrada e retorne uma nova lista contendo apenas os elementos que aparecem em pelo menos uma das listas, mas não em ambas.

```python
def elementos_unicos(lista1, lista2):
    """
    Recebe duas listas de números inteiros e retorna uma nova lista
    contendo apenas os elementos que aparecem em pelo menos uma das listas,
    mas não em ambas.
    """
    set1 = set(lista1)
    set2 = set(lista2)
    unicos = set1.symmetric_difference(set2)
    return list(unicos)
```

```python
lista1 = [1, 2, 3, 4, 5, 6, 7]
lista2 = [4, 5, 6, 7, 8, 9, 10]
unicos = elementos_unicos(lista1, lista2)
print(unicos)
```

### 13. Crie um programa em Python que receba duas listas de números inteiros como entrada e retorne uma nova lista contendo apenas os elementos que aparecem em ambas as listas, mas sem repetições.

```python
def elementos_comuns(lista1, lista2):
    """
    Recebe duas listas de números inteiros e retorna uma nova lista
    contendo apenas os elementos que aparecem em ambas as listas, sem repetições.
    """
    set1 = set(lista1)
    set2 = set(lista2)
    comuns = set1.intersection(set2)
    return list(comuns)
```

```python
lista1 = [1, 2, 3, 4, 5]
lista2 = [4, 5, 6, 7, 8]
comuns = elementos_comuns(lista1, lista2)
print(comuns)
```

### 14. Crie um programa em Python que receba uma lista de dicionários como entrada, onde cada dicionário representa um produto com nome e preço, e retorne o nome do produto mais caro.

```python
produtos = [{"nome": "celular", "preco": 1500},
            {"nome": "notebook", "preco": 3500},
            {"nome": "tablet", "preco": 1200},
            {"nome": "fones de ouvido", "preco": 200},
            {"nome": "smartwatch", "preco": 800}]
```

```python
def produto_mais_caro(produtos):
    """
    Recebe uma lista de dicionários representando produtos com nome e preço e
    retorna o nome do produto mais caro.
    """
    mais_caro = None
    preco_mais_caro = 0
    for produto in produtos:
        if produto["preco"] > preco_mais_caro:
            preco_mais_caro = produto["preco"]
            mais_caro = produto["nome"]
    return mais_caro
```

```python
produtos = [{"nome": "celular", "preco": 1500},
            {"nome": "notebook", "preco": 3500},
            {"nome": "tablet", "preco": 1200},
            {"nome": "fones de ouvido", "preco": 200},
            {"nome": "smartwatch", "preco": 800}]
mais_caro = produto_mais_caro(produtos)
print(mais_caro)
```

### 15. Crie um programa em Python que receba um dicionário como entrada, onde as chaves representam nomes de produtos e os valores representam as quantidades disponíveis em estoque, e retorne o nome do produto com estoque mais baixo.

```python
def produto_com_menor_estoque(estoque):
    """
    Recebe um dicionário representando o estoque de produtos, onde as chaves
    são os nomes dos produtos e os valores são as quantidades disponíveis, e
    retorna o nome do produto com menor quantidade em estoque.
    """
    menor_estoque = None
    quantidade_menor_estoque = float("inf")
    for produto, quantidade in estoque.items():
        if quantidade < quantidade_menor_estoque:
            quantidade_menor_estoque = quantidade
            menor_estoque = produto
    return menor_estoque
```

```python
estoque = {"celular": 50, "notebook": 20, "tablet": 100, "fones de ouvido": 5, "smartwatch": 15}
menor_estoque = produto_com_menor_estoque(estoque)
print(menor_estoque)
```
