### 16. Crie um programa em Python que receba uma lista de dicionários como entrada, onde cada dicionário representa um aluno com nome e notas em diferentes disciplinas, e retorne o nome do aluno com a maior média ponderada, sendo que a média ponderada deve ser calculada utilizando as notas como pesos.
### media_ponderada = (8.5x2 + 9.0x3 + 7.0x2)/7

```python
alunos = [{"nome": "João", "notas": {"Matemática": 8.5, "Português": 9.0, "História": 7.0}},
          {"nome": "Maria", "notas": {"Matemática": 9.5, "Português": 8.5, "História": 6.0}},
          {"nome": "Pedro", "notas": {"Matemática": 7.0, "Português": 6.5, "História": 9.0}},
          {"nome": "Ana", "notas": {"Matemática": 8.0, "Português": 7.5, "História": 8.0}}]
```

```python
def calcular_media_ponderada(notas):
    return sum([nota['nota'] * nota['peso'] for nota in notas])/sum([nota['peso'] for nota in notas])

def aluno_com_maior_media(alunos):
    maior_media = 0
    nome_maior_media = ""
    for aluno in alunos:
        media_ponderada = calcular_media_ponderada(aluno['notas'])
        if media_ponderada > maior_media:
            maior_media = media_ponderada
            nome_maior_media = aluno['nome']
    return nome_maior_media
```

```python
alunos = [
    {
        'nome': 'João',
        'notas': [
            {'nota': 8.5, 'peso': 2},
            {'nota': 9.0, 'peso': 3},
            {'nota': 7.0, 'peso': 2},
        ]
    },
    {
        'nome': 'Maria',
        'notas': [
            {'nota': 9.5, 'peso': 2},
            {'nota': 8.5, 'peso': 3},
            {'nota': 6.0, 'peso': 2},
        ]
    },
    {
        'nome': 'Pedro',
        'notas': [
            {'nota': 7.0, 'peso': 2},
            {'nota': 6.5, 'peso': 3},
            {'nota': 9.0, 'peso': 2},
        ]
    },
    {
        'nome': 'Ana',
        'notas': [
            {'nota': 8.0, 'peso': 2},
            {'nota': 7.5, 'peso': 3},
            {'nota': 8.0, 'peso': 2},
        ]
    },
]

print(aluno_com_maior_media(alunos))
```

### 17. Crie um programa em Python que receba três números como entrada e imprima "Maior" se o primeiro número for maior do que a soma dos outros dois, "Menor" se o primeiro número for menor do que a soma dos outros dois, ou "Igual" caso contrário.

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
num3 = float(input("Digite o terceiro número: "))
```

```python
def comparar_numeros(num1, num2, num3):
    if num1 > num2 + num3:
        print(f"O número {num1} é MAIOR que a soma de {num2} + {num3}")
    elif num1 < num2 + num3:
        print(f"O número {num1} é MENOR que a soma de {num2} + {num3}")
    else:
        print(f"O número {num1} é IGUAL a soma dos números {num2} + {num3}")
comparar_numeros(num1, num2, num3)
```

### 18. Crie um programa em Python que receba a idade e a altura de uma pessoa como entrada e classifique-a de acordo com as seguintes regras: se a idade for menor do que 18 anos e a altura for menor do que 1,60 m, imprimir "Adolescente Baixo"; se a idade for menor do que 18 anos e a altura for maior do que 1,60 m, imprimir "Adolescente Alto"; se a idade for maior ou igual a 18 anos e a altura for menor do que 1,60 m, imprimir "Adulto Baixo"; se a idade for maior ou igual a 18 anos e a altura for maior do que 1,60 m, imprimir "Adulto Alto".

```python
idade = int(input("Digite a idade: "))
altura = float(input("Digite a altura: "))
```

```python
def classificar_pessoa(idade, altura):
    if idade < 18 and altura < 1.6:
        print("Adolescente Baixo")
    elif idade < 18 and altura >= 1.6:
        print("Adolescente Alto")
    elif idade >= 18 and altura < 1.6:
        print("Adulto Baixo")
    else:
        print("Adulto Alto")
classificar_pessoa(idade, altura)
```

### 19. Crie um programa em Python que receba uma string como entrada e imprima "Palíndromo" se a string for um palíndromo (ou seja, se a string lida de trás para frente for igual à string original), ou "Não Palíndromo" caso contrário.

```python
texto = input("Digite um texto: ")
```

```python
def eh_palindromo(texto):
    texto = texto.replace(' ', '').lower() # Remove espaços em branco e converte para minúsculas
    if texto == texto[::-1]: # Compara o texto original com o texto invertida
        return "Palíndromo"
    else:
        return "Não Palíndromo"
eh_palindromo(texto)
```

### 20. Crie um programa em Python que receba uma lista de números inteiros como entrada e calcule a soma dos quadrados dos números ímpares da lista.

```python
def soma_quadrados_impares(lista):
    soma = 0
    for num in lista:
        if num % 2 != 0: # Verifica se o número é ímpar
            soma += num**2 # Adiciona o quadrado do número ímpar à soma
    return soma

lista = [1, 2, 3, 4, 5]
resultado = soma_quadrados_impares(lista)
print(resultado) 
```
