### 21. Crie um programa em Python que receba uma string como entrada e imprima a string sem as vogais.

```python
def remove_vogais(texto):
    vogais = "aeiouAEIOU" # Definimos uma string com as vogais maiúsculas e minúsculas
    texto_sem_vogais = "" # Inicializamos uma variável vazia para armazenar o resultado
    
    for letra in texto:
        if letra not in vogais: # Se a letra não é uma vogal, adicionamos ela à string de resultado
            texto_sem_vogais += letra
    
    print(texto_sem_vogais)

texto = "E se fosse tudo tão grande quanto o oceano."
remove_vogais(texto) 
```

### 22. Crie um programa em Python que imprima a soma dos n primeiros números primos, onde n é um número fornecido pelo usuário.

```python
n = int(input("Digite um número inteiro positivo: "))
```

```python
def eh_primo(numero):
    if numero < 2: # Números menores que 2 não são primos
        return False
    for i in range(2, int(numero**0.5) + 1): # Percorre os possíveis divisores do número
        if numero % i == 0: # Se o número for divisível por i, não é primo
            return False
    return True

def soma_n_primos(n):
    soma = 0
    numero = 2 # Começamos a busca pelos números primos em 2
    
    while n > 0:
        if eh_primo(numero):
            soma += numero
            n -= 1
        numero += 1
        
    return soma

resultado = soma_n_primos(n)
print("A soma dos {} primeiros números primos é: {}".format(n, resultado))
```

### numeros = [2, 5, 9, 12, 18, 23, 27, 30, 35, 40]

```python
def soma_pares_impares(lista):
    soma_pares = 0
    soma_impares = 0
    
    for num in lista:
        if num % 2 == 0: # Verifica se o número é par
            soma_pares += num
        else:
            soma_impares += num
    
    print("A soma dos números pares é: {}".format(soma_pares))
    print("A soma dos números ímpares é: {}".format(soma_impares))

lista = [2, 5, 9, 12, 18, 23, 27, 30, 35, 40]
soma_pares_impares(lista) 
```

### 25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).

```python
num = int(input("Digite um número inteiro positivo: "))
```

```python
def eh_numero_perfeito(num):
    soma_divisores = 0
    for i in range(1, num):
        if num % i == 0:
            soma_divisores += i
    if soma_divisores == num:
        return True
    else:
        return False

if eh_numero_perfeito(num):
    print("{} é um número perfeito!".format(num))
else:
    print("{} não é um número perfeito.".format(num))
```

### 26. Crie um programa em Python que calcule a série de Fibonacci para um determinado número de termos fornecido pelo usuário. A série de Fibonacci é uma sequência em que cada número é a soma dos dois números anteriores: 0, 1, 1, 2, 3, 5, 8, 13, ...

```python
n = int(input("Digite o número de termos desejado na sequência de Fibonacci: "))
```

```python
def fibonacci(n):
    if n == 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        seq = [0, 1]
        while len(seq) < n:
            seq.append(seq[-1] + seq[-2])
        return seq

seq_fibonacci = fibonacci(n)
print("A sequência de Fibonacci para os {} termos é: {}".format(n, seq_fibonacci))
```
